```python
X = data.drop("country",axis=1)
```

```python
scaler = MinMaxScaler()
```

```python
X=scaler.fit_transform(X)
```

```python
wcss = []
for i in range(1,8):
    kmeans = KMeans(n_clusters=i,random_state=11)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,8),wcss)
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
data['result'] = y_kmeans
```

```python
features = data[['child_mort', 'exports', 'health', 'imports', 'income', 'inflation',
                 'life_expec', 'total_fer', 'gdpp']]
```

```python
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(features_scaled)
```

```python
hierarchical_clusters = linkage(features_scaled, method='ward')
```

```python
plt.figure(figsize=(12, 8))
dendrogram(hierarchical_clusters, labels=data['country'].tolist(), orientation='top', leaf_font_size=10)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Countries')
plt.ylabel('Distance')
plt.show()
```
